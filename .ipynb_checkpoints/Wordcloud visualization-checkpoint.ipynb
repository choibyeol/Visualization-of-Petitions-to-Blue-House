```python
#엑셀을 읽기 위한 준비 과정
from openpyxl import load_workbook

read_wb = load_workbook("./bluehouse.xlsx")
read_ws = read_wb.active

list_excel = []

for i in range(1, 151):
    print(read_ws.cell(i, 1).value.strip())
    list_excel.append(read_ws.cell(i, 1).value.strip()) 
```

```python
from konlpy.tag import Kkma
import collections
kkma = Kkma()

list_temp = []

for row in list_excel:
    list_temp = kkma.nouns(row) + list_temp
    
print(list_temp)
```

```python
list_result = []
for check in list_temp:
    if len(check) > 1:
        print(check)
        list_result.append(check)
        
list_result
```

```python
#wordcloud에 넣기 위한 문자열
last_text = ""
```

```python
for i in list_result:
    last_text = last_text + " " + i
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
wordcloud = WordCloud(font_path="C:\\WINDOWS\\Fonts\\NanumGothicExtraBold.ttf").generate(last_text)
```

```python
wordcloud.words_
```

```python
plt.figure(figsize = (12, 12))
plt.imshow(wordcloud, interpolation = "bilinear")
plt.axis("off")
plt.show()
```

```python
from os import path
from PIL import Image
import numpy as np
import os

d = path.dirname(__file__) if "__file__" in locals() else os.getcwd()
```

```python
wordcloud = WordCloud(
    font_path = "C:\\WINDOWS\\Fonts\\NanumGothicExtraBold.ttf",
    background_color = "white",
    mask = mask,
).generate(last_text)
```
