```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
list_data = [('부동산', 15),
 ('반대', 14),
 ('서울', 12),
 ('박원', 11),
 ('시장', 11),
 ('청원', 11),
 ('대책', 11),
 ('처벌', 10),
 ('소급', 10),
 ('적용', 10),
 ('요청', 8),
 ('철회', 8),
 ('소급적용', 8),
 ('박원순', 7),
 ('국토', 6),
 ('관련', 6),
 ('임대', 6),
 ('주택', 6),
 ('특별', 5),
 ('사건', 5)]
```

```python
list_string = []
list_number = []
```

```python
for s, n in list_data:
    print(s, n)
    list_string.append(s)
    list_number.append(n)
```

```python
label = list_string
index = np.arange(len(label))
plt.bar(index, list_number)

#폰트 설정, 한글 깨지는 문제 해결
plt.rcParams["font.family"] = 'nanummyeongjo'

#그리기
plt.title('my title')
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.xticks(index, label)
plt.xticks(rotation = 55)

plt.show()
```
